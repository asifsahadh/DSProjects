## Data Pre-Processing

```python
# converting index data type and column data type to list
df_can.index = df_can.index.tolist()
df_can.columns = df_can.columns.tolist()

print(type(df_can.index))
print(type(df_can.columns))
```

```python
df_can.shape
```

```python
# dropping unwanted columns
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)
df_can.head()
```

```python
# renaming columns
df_can.rename(columns = {'OdName' : 'Country', 'AreaName' : 'Continent', 'RegName' : 'Region'}, inplace = True)
df_can.columns
```

```python
# creating a new feature 'Total' which encapsulates the total number of immigrations from 1980 to 2013
df_can['Total'] = df_can[range(1980, 2014)].sum(axis = 1)
df_can['Total']
```

```python
# setting the country names as index for the data frame
df_can.set_index('Country', inplace = True)
df_can.index.name = None
df_can.head()
```

```python
df_can.columns = list(map(str, df_can.columns))
df_can.head()
```

```python
df_can.sort_values('Total', ascending = False, inplace = True)
```
