```python
#importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#loading the dataset
df = pd.read_excel(r"C:\Users\asifs\Desktop\DS Projects\Amazon Sales Data\amazon_sales_data_transformed.xlsx")
pd.set_option('display.max.rows', 101)
df
```

```python
#checking the unique values in 'Item Type' column
df['Item Type'].unique()
```

```python
#viewing correlation between unit price and units sold
sns.regplot(x = df['Unit Price'], y = df['Units Sold'], scatter_kws = {'color' : 'red'}, line_kws = {'color' : 'blue'})
plt.title('Unit Price vs Units Sold')
plt.show()
```

```python
#viewing correlation matrix for numeric values in heatmap form
correlation_matrix = df.corr()
plt.figure(figsize = (8, 6))
sns.heatmap(correlation_matrix, annot = True)
plt.show
```

```python
#converting all the attributes to numerical format
df_numerical = df
for col_name in df_numerical.columns:
  if df_numerical[col_name].dtype == 'object':
    df_numerical[col_name] = df_numerical[col_name].astype('category')
    df_numerical[col_name] = df_numerical[col_name].cat.codes
```

```python
#viewing correlation matrix of the whole dataset in heatmap form
correlation_matrix_numerical = df_numerical.corr()
plt.figure(figsize = (14, 12))
sns.heatmap(correlation_matrix_numerical, annot = True)
plt.show()
```
