```python
#importing libraries
import pandas as pd
```

```python
#extracting and viewing the dataset
df = pd.read_csv(r"C:\Users\asifs\Desktop\DS Projects\Amazon Sales Data\Amazon Sales Data.csv")
pd.set_option('display.max.rows', 100)
df
```

```python
#checking for null values
df.isnull().sum()
```

```python
#dropping duplicate rows if any
df.drop_duplicates()
```

```python
#checking data types
df.dtypes
```

```python
#fixing date data type
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df
```

```python
#creating sale month and sale year column
df['Sale Month'] = df['Order Date'].dt.month
df['Sale Year'] = df['Order Date'].dt.year
df
```

```python
#viewing sale month values
df['Sale Month'].unique()
```

```python
#naming sale month column
month_dict = {'1' : 'January', '2' : 'February', '3' : 'March', '4' : 'April', '5' : 'May', '6' : 'June', '7' : 'July', '8' : 'August', '9' : 'September', '10' : 'October', '11' : 'November', '12' : 'December'}
df['Sale Month'] = df['Sale Month'].astype(str)
df['Sale Month'] = df['Sale Month'].map(month_dict)
df
```

```python
#creating shipping duration column
df['Shipping Duration'] = df['Ship Date'] - df['Order Date']
df
```

```python
#sorting dataframe
df.sort_values(by = ['Region', 'Country'], ascending = True, inplace = True)
df.reset_index(inplace = True)
df
```

```python
#removing index column
df.drop(columns = ['index'], inplace = True)
df
```

```python
#changing order priority data
df['Order Priority'] = df['Order Priority'].replace('L', 'Low')
df['Order Priority'] = df['Order Priority'].replace('M', 'Medium')
df['Order Priority'] = df['Order Priority'].replace('H', 'High')
df['Order Priority'] = df['Order Priority'].replace('C', 'Critical')
df
```

```python
#loading the transformed dataframe
transformed_df_path = r"C:\Users\asifs\Desktop\DS Projects\Amazon Sales Data\amazon_sales_data_transformed.xlsx"
df.to_excel(transformed_df_path, index = False)
```
